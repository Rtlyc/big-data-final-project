```python
import dask.dataframe as dd
filepath = "dataset/Arts.txt"
df = dd.read_csv(filepath, delimiter="\t")
df.head(20)
```

```python
with open(filepath) as f:
    lines = [line.rstrip('\n') for line in f]
print(lines[:10])
# columns = []
# for line in lines[:8]:
#     key, value = line.split(":")
#     columns.append(key)
# print(columns)
```

```python
from collections import defaultdict
ref=defaultdict(list)

for line in lines:
    if not line: continue
    colonPos = line.find(":")
    key = line[:colonPos]    
    value = line[colonPos+2:]
    ref[key].append(value)
```

```python
df = dd.from_pandas(pd.DataFrame(data=ref), npartitions=1)
df.head()
```
